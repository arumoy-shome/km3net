```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
hits = pd.read_csv("../data/hits.csv")
hits = hits.sort_values(by=['time'])
hits
```

# Load the data

```python
slices = list(range(0, 10000000, 15000))

timeslices = []
while len(timeslices) < 10:
    choice = random.choice(slices)
    if choice < slices[-1]:
        timeslices.append((choice, choice+15000))

timeslices
```

```python
fig, axs = plt.subplots(5, 2, figsize=(15, 15))

# plot 1
timeslices_idx = 0
slice = hits[(hits.time >= timeslices[timeslices_idx][0]) & (hits.time < timeslices[timeslices_idx][1])]
axs[0, 0].scatter(slice.time, slice.pos_x)

# plot 2
timeslices_idx = 1
slice = hits[(hits.time >= timeslices[timeslices_idx][0]) & (hits.time < timeslices[timeslices_idx][1])]
axs[0, 1].scatter(slice.time, slice.pos_x)

# plot 3
timeslices_idx = 2
slice = hits[(hits.time >= timeslices[timeslices_idx][0]) & (hits.time < timeslices[timeslices_idx][1])]
axs[1, 0].scatter(slice.time, slice.pos_x)

# plot 4
timeslices_idx = 3
slice = hits[(hits.time >= timeslices[timeslices_idx][0]) & (hits.time < timeslices[timeslices_idx][1])]
axs[1, 1].scatter(slice.time, slice.pos_x)

# plot 5
timeslices_idx = 4
slice = hits[(hits.time >= timeslices[timeslices_idx][0]) & (hits.time < timeslices[timeslices_idx][1])]
axs[2, 0].scatter(slice.time, slice.pos_x)
    
# plot 6
timeslices_idx = 5
slice = hits[(hits.time >= timeslices[timeslices_idx][0]) & (hits.time < timeslices[timeslices_idx][1])]
axs[2, 1].scatter(slice.time, slice.pos_x)

# plot 7
timeslices_idx = 6
slice = hits[(hits.time >= timeslices[timeslices_idx][0]) & (hits.time < timeslices[timeslices_idx][1])]
axs[3, 0].scatter(slice.time, slice.pos_x)

# plot 8
timeslices_idx = 7
slice = hits[(hits.time >= timeslices[timeslices_idx][0]) & (hits.time < timeslices[timeslices_idx][1])]
axs[3, 1].scatter(slice.time, slice.pos_x)

# plot 9
timeslices_idx = 8
slice = hits[(hits.time >= timeslices[timeslices_idx][0]) & (hits.time < timeslices[timeslices_idx][1])]
axs[4, 0].scatter(slice.time, slice.pos_x)

# plot 10
timeslices_idx = 9
slice = hits[(hits.time >= timeslices[timeslices_idx][0]) & (hits.time < timeslices[timeslices_idx][1])]
axs[4, 1].scatter(slice.time, slice.pos_x)
```
